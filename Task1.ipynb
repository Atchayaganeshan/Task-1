```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
n_samples = 750
```

```python
data = {
    'High_GPA': np.round(np.random.uniform(2.0, 4.0, n_samples), 2),
    'Extracurriculars': np.random.randint(0, 10, n_samples),
    'Parental_Income': np.random.randint(20000, 150000, n_samples),
    'Hours_Studied': np.random.randint(1, 40, n_samples),
    'Entrance_Test_Score': np.round(np.random.uniform(50, 100, n_samples), 2),
    'Distance_from_College': np.round(np.random.uniform(0.5, 50.0, n_samples), 2),
    'Internet_Availability': np.random.choice(['Yes', 'No'], n_samples)
}
```

```python
df = pd.DataFrame(data)
```

```python
internet_binary = df['Internet_Availability'].map({'Yes': 1, 'No': 0})
```

```python
pass_prob = (
    0.3 * df['High_GPA'] +
    0.2 * df['Hours_Studied'] +
    0.2 * (df['Entrance_Test_Score'] / 10) +
    0.1 * internet_binary +
    np.random.normal(0, 0.5, n_samples)
)
```

```python
df['Final_Result'] = np.where(pass_prob > 4.5, 'Pass', 'Fail')
```

```python
df.to_csv('student_pass_fail_dataset.csv', index=False)
print("Dataset generated and saved as 'student_pass_fail_dataset.csv'")
```
